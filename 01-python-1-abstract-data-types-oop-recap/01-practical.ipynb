### Polymorphism
**Question 12:** `Language` class has no class attributes, but a class method called `good_morning`. `Maori`, `Japanese` & `German` class inherit from `Language` class. When you execute the following code, what is happening & why is it happening? Refactor the code to display the expected output.

```python
# Write your solution here

class Language:
    def good_morning(self):
        raise NotImplementedError

class Maori(Language):
    def good_morning(self):
        print('Morena')

class Japanese(Language):
    pass

class German(Language):
    def good_morning(self):
        print('Guten Morgen')

def main():
    maori = Maori()
    japanese = Japanese()
    german = German()
    for lang in (maori, japanese, german):
        lang.good_morning()

if __name__ == '__main__':
    main()

# Expected output:

# Morena
# おはようございます
# Guten Morgen
```
